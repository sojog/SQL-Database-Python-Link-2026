```python
from dotenv import load_dotenv
import os

# load .env file to environment
load_dotenv()

SECRET_USERNAME = os.getenv('SECRET_USERNAME')
SECRET_PASSWORD = os.getenv('SECRET_PASSWORD')
```

```python

HOSTNAME = '127.0.0.1'
DATABASE_NAME = 'Facultate'
```

```python
from mysql.connector import (connection)

cnx = connection.MySQLConnection(user=SECRET_USERNAME, password=SECRET_PASSWORD,
                                 host=HOSTNAME,
                                 database=DATABASE_NAME)
# cnx.close()
```

```python
import mysql.connector

cnx = mysql.connector.connect(user=SECRET_USERNAME, password=SECRET_PASSWORD,
                              host=HOSTNAME,
                              database=DATABASE_NAME)
# 
```

```python
SQL_QUERY = "SELECT * FROM `Facultate`.`Student`;"

pd.read_sql(SQL_QUERY, cnx)
```
